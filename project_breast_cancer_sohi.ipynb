###  Часть 3. Предобработка данных 

Представим переменную diagnosis в бинарном виде

```python
df.diagnosis = df.diagnosis.replace({'M': 1, 'B': 0})
df.head()
```

Удалим столбец с пустыми значениями и идетификаторы пациентов

```python
df.drop(columns=['Unnamed: 32', 'id'], inplace=True)
```

```python
def divide_by_scale(df, target):
    lower_1 = []
    higher_1 = []
    normal = []
    big = []
    for column in df.drop(target, axis=1).columns:
        if np.mean(df[column]) < 0.1:
            lower_1.append(column)
        elif np.mean(df[column]) > 0.1 and np.mean(df[column]) < 1:
            higher_1.append(column)
        elif np.mean(df[column]) > 1 and np.mean(df[column]) < 50:
            normal.append(column)
        else:
            big.append(column)
    return lower_1, higher_1, normal, big
```

```python
def divide_by_sep(df, sep):
    columns_sep = []
    for column in df.columns:
        if sep in column:
            columns_sep.append(column)
    return columns_sep
```

```python
mean_columns = divide_by_sep(df, '_mean')
se_columns = divide_by_sep(df, '_se')
worst_columns = divide_by_sep(df, '_worst')
```

```python
lower_1, higher_1, normal, big = divide_by_scale(df, 'diagnosis')
```

###  Отбор признаков

Судя по графикам, признаки группы _se оказались наимение информативными. Это может негативно сказаться на качестве обучения модели, поэтому мы от них избавимся

Остальные признаки (_mean, _worst) мы также преобразуем, взяв среднее арифметическое между средним (_mean) значением каждого признака и наибольшим (_worst)

```python
def format_columns(df):
    new_df = df.copy()
    for column in df.columns:
        pure_column = column.split('_')[0]
        if pure_column not in new_df.columns:
            new_df[pure_column] = (new_df['{0}_mean'. format(pure_column)] + new_df['{0}_worst'. format(pure_column)]) / 2
            new_df.drop(columns=['{0}_mean'. format(pure_column), 
                                 '{0}_worst'. format(pure_column), '{0}_se'. format(pure_column)], inplace=True)
    return new_df
```

```python
df.rename(columns={'fractal_dimension_mean': 'fractaldimension_mean',
                  'fractal_dimension_worst': 'fractaldimension_worst',
                  'fractal_dimension_se': 'fractaldimension_se'}, inplace=True)
```

###  Часть 7. Построение кривых валидации и обучения 

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    '''
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_learning_curve.html
    '''
    
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='roc_auc')
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

Обучим модели с наилучшими параметрами

```python
gbm = GradientBoostingClassifier(loss='deviance', learning_rate=1.15, min_impurity_decrease=0.1,\
                                 max_depth=2, max_features='sqrt', n_estimators=100, warm_start=True,\
                                 min_samples_leaf=1, min_samples_split=2, random_state=21)
gbm.fit(X_train, y_train)
```

```python
logit = LogisticRegression(C=0.7, dual=False,
                fit_intercept=True,
                intercept_scaling=4,
                max_iter=100,
                solver='liblinear',
                random_state=21,
                warm_start=False).fit(X_train, y_train)
```

```python
plt.figure(figsize=(8, 6))
plot_learning_curve(rfc, 'Random Forest', X_train, y_train, cv=3, n_jobs=-1);
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plot_learning_curve(logit, 'Logistic Regression', X_train, y_train, cv=3, n_jobs=-1);
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plot_learning_curve(gbm, 'Gradient Boosting', X_train, y_train, cv=3, n_jobs=-1);
plt.show()
```

###  Часть 8. Прогноз для тестовой или отложенной выборки и оценка модели с описанием выбранной метрики

В качестве метрики целесообразно будет использовать precision или recall, т.к. в данных присутствует дисбаланс классов
Учитывая специфику задачи, будем использовать recall, потому что неверная классификация больных со злокачественной опухолью будет наиболее критична

Прогноз случайного леса:

```python
predictions = rfc.predict(X_test)
round(recall_score(predictions, y_test), 3)
```

Градиентный бустинг:

```python
predictions = gbm.predict(X_test)
round(recall_score(predictions, y_test), 3)
```

Логистическая регрессия:

```python
predictions = logit.predict(X_test)
round(recall_score(predictions, y_test), 3)
```
